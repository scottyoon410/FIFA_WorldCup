# Then, Who is 'Underdog'?

```python
# Create a pivot table to count the number of wins per team
win_counts = pd.pivot_table(merged_df, values='Award Id', index='Home Team Name', aggfunc='count')

# Calculate the 30th percentile of win counts
win_count_30pct = win_counts.quantile(0.3)[0]

# Filter the win_counts dataframe to only include the top 5 teams with the highest win counts
top_teams = win_counts.nlargest(5, 'Award Id').index.tolist()

# Filter the merged dataframe to only include matches where a top team played
top_team_matches = merged_df[(merged_df['Home Team Name'].isin(top_teams)) | (merged_df['Away Team Name'].isin(top_teams))]
```

```python
# Create a countplot of top teams by tournament without hue
sns.countplot(x='Tournament Name', data=top_team_matches)
plt.xticks(rotation=90)
plt.title('Number of Matches Played by Underdog Teams by Tournament')
plt.show()
```

```python
# Create a scatterplot of top teams' scores by tournament without hue
sns.scatterplot(x='Tournament Name', y='Home Team Score', data=top_team_matches)
plt.xticks(rotation=90)
plt.title('Scores of Underdog Teams by Tournament')
plt.show()
```

```python
# Create a barplot of top teams' score margins by tournament without hue
sns.barplot(x='Tournament Name', y='Home Team Score Margin', data=top_team_matches)
plt.xticks(rotation=90)
plt.title('Score Margins of Underdog Teams by Tournament')
plt.show()
```

Score Margins of Underdog Teams by Tournament에서 홈팀 점수 마진이 -0.5골 차이의 마진을 졌다는 의미이다. 즉, 상위 팀은 평균적으로 상대방보다 적은 골을 넣었다는 의미이다. 

따라서 2014 브라질 월드컵에서 어떤 측면이 UnderDog의 우승 횟수를 늘릴 수 있었는지 살펴 볼 예정이다.

```python
# 2014 FIFA World Cup 대회에 대해서만 UnderDog 우승을 위한 요인 찾기

# UnderDog 팀의 이름
underdog_team = "UnderDog"

# 2014 FIFA World Cup 대회에 대한 데이터만 추출
df_wc2014 = merged_df[merged_df["Tournament Name"] == "2014 FIFA World Cup"]

print(df_wc2014.info())
```

```python
# UnderDog 팀의 승리 기록
underdog_wins = len(df_wc2014[(df_wc2014["Home Team Name"] == underdog_team) & (df_wc2014["Home Team Score"] > df_wc2014["Away Team Score"])] 
                    + df_wc2014[(df_wc2014["Away Team Name"] == underdog_team) & (df_wc2014["Away Team Score"] > df_wc2014["Home Team Score"])])

# 나머지 팀들 중 UnderDog 팀의 승리보다 많은 승리를 기록한 팀들의 수
better_teams_count = len(df_wc2014[(df_wc2014["Home Team Name"] != underdog_team) & (df_wc2014["Home Team Score"] > underdog_wins)] 
                          + df_wc2014[(df_wc2014["Away Team Name"] != underdog_team) & (df_wc2014["Away Team Score"] > underdog_wins)])
```

```python
# UnderDog 팀이 더 많은 경기를 이겼을 경우
if underdog_wins > better_teams_count:
    print(f"The underdog team {underdog_team} won more games ({underdog_wins}) than other teams ({better_teams_count}) in the tournament!")
else:
    print(f"The underdog team {underdog_team} did not win more games ({underdog_wins}) than other teams ({better_teams_count}) in the tournament.")
```

```python
df_wc2014["Underdog Wins More"] = df_wc2014.apply(lambda row: 1 if (row["Home Team Name"] == underdog_team and row["Home Team Score"] > row["Away Team Score"] and row["Home Team Score"] > underdog_wins) or (row["Away Team Name"] == underdog_team and row["Away Team Score"] > row["Home Team Score"] and row["Away Team Score"] > underdog_wins) else 0, axis=1)

underdog_win_more = len(df_wc2014[df_wc2014['Underdog Wins More'] == True])
underdog_win_less = len(df_wc2014[df_wc2014['Underdog Wins More'] == False])

labels = ['Underdog Wins More', 'Underdog Wins Less']
sizes = [underdog_win_more, underdog_win_less]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#FF6347','#ADD8E6'])
plt.axis('equal')
plt.title('Underdog Win Percentage')
plt.show()
```
